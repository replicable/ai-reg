```python
import openai, numpy as np
import matplotlib.pyplot as plt

words = ['emotional', 'smart', 'cool', 'hysterical', 'married', 'rational', 'poor',
         'reliable', 'brave', 'sexy', 'strong', 'aggressive', 'handsome', 'successful']
results = {'he': list(), 'she': list()}

for word in words:
  for gender, scores in results.items():
    resp = openai.Embedding.create(input=[word, gender], engine="text-similarity-davinci-001")
    similarity_score = np.dot(resp['data'][0]['embedding'], resp['data'][1]['embedding'])
    scores.append(similarity_score)    
fig, ax = plt.subplots()
ax.scatter(results['he'], results['she'])
for i, word in enumerate(words):
    ax.annotate(word, (results['he'][i], results['she'][i]))

ax.plot([0, 1], [0, 1], color='red', transform=ax.transAxes)
plt.xlabel("he")
plt.ylabel("she")
plt.show()
```
